# Defining Selenium Locator on WEB<br>
source: [Selenium Locator strategies](https://www.selenium.dev/documentation/webdriver/elements/locators/)

The Selenium locators we use for now is `xpath` strategies. It can be found on previous documentation or using Chrome Extension [Ranorex Selocity](https://chrome.google.com/webstore/detail/ranorex-selocity/ocgghcnnjekfpbmafindjmijdpopafoe/related?hl=en-US) to help us define locators quicker<br>
`ID` : is the best locators for Selenium, but not all elements are using ID
<br>


## Obtaining the Locators

### Ranorex method

#### Obtain the xpath
- Right click on the desired elements (`username` textbox)
- Select Ranorex Selocity
- Select any available method (on this case: `CSS, xpath, rxpath`)
<br>
- Try to paste the result to validate the path (`/html//input[@id='username']`)

#### Verify the xpath on Chrome Devtools
- Switch to Console tab
- Type `$x("")` on the next command line

- Paste the obtained xpath before inside the double quotes -> `$x("/html//input[@id='username'")`

- Push Enter to obtain the correct path
<br>
By executing this validation we only receive one result which means, it's valid path and pointing at our destination correctly (hovering mouse pointers on the result)

This method is mostly used when we receiving more than elements on our xpath
