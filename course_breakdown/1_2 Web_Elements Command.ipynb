The `current_url` [method](https://www.geeksforgeeks.org/how-to-get-current_url-using-selenium-in-python/) is used to retrieve the URL of the webpage the user is currently accessing. It takes no argument & gives the URL of the current webpage loaded by the driver in selenium. It only can be assigned into variable and we can print the variable to ensure the result later
