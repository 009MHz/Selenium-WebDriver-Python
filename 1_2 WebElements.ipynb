## [Find Elements](https://selenium-python.readthedocs.io/locating-elements.html)
Is the function on the python under ***Selenium - Webdriver*** package that used to find the HTML elements under tested page. This module function receiving some values. Below are list of several attributes that can be passed under the package function

- Find Elements accepted attributes:
    - ID = "id"
    - NAME = "name"
    - XPATH = "xpath"
    - LINK_TEXT = "link text"
    - PARTIAL_LINK_TEXT = "partial link text"
    - TAG_NAME = "tag name"
    - CLASS_NAME = "class name"
    - CSS_SELECTOR = "css selector"

- Find Elements function use:
    - `find_element(By.ID, "id")`
    - `find_element(By.NAME, "name")`
    - `find_element(By.XPATH, "xpath")`
    - `find_element(By.LINK_TEXT, "link text")`
    - `find_element(By.PARTIAL_LINK_TEXT, "partial link text")`
    - `find_element(By.TAG_NAME, "tag name")`
    - `find_element(By.CLASS_NAME, "class name")`
    - `find_element(By.CSS_SELECTOR, "css selector")`
