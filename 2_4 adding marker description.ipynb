# pytest.ini
is a configuration file for the `pytest` testing framework. It can be used to specify configuration options for `pytest` that affect how it runs. For example, you can use it to specify the directories in which `pytest` should search for test files, or to specify which tests should be run when pytest is run.

When we run the test without adding this configuration, we'll receive an error warning on the console log since the `pytest marker` has no description on it although the test is pass

2. Add `.ini` file to the directory
