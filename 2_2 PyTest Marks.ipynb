# PyTest Marks

## Definition
Basically, `PyTest` ***Marks*** is a label to filter the PyTest execution from the scripts. 

`pytest.mark` is a decorator that allows you to add metadata to test functions. This metadata can be used to include or exclude test functions from a test run, or to apply additional behavior to the test function

For Example, we can add our `TestPositiveLogin` scenario on `test_login_page.py` with these labels:
 <br>
`login`, & `positive`

## Using the marker
To use the marker, we can pass these scripts on the first line before the function is executed <br>
`@pytest.mark.label_name`

# Run Test on the marker

## 1. Run the test on terminal
- Navigate to the terminal
- Type this script on the terminal
`pytest -m {marker_name} -v`
    - `pytest` : run test cases that are written using the pytest framework
    - `-m {marker_name}` : option is used to filter the test cases that will be run, based on a marker
    - `-v` : option is used to enable verbose output, which will print more detailed information about the test run to the console






## 2.Run the Test on VS Code Tests Pannel
1. Configure the Tests on from the pannel

2. Find the script that will be tested

3. Run the test from **Test Explorer**

4. See the logs for the result


## 2. Run the Test on PyCharm
1. Navigate to Run/Debug Configuration

2. Add new pytest Configuration

3. Change the test name & modify the Additional Arguments with this value: <br>
`-m {marker name}` <br>

4. Select created configuration to run the test

5. Check the console result
