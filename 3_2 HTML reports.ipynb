# [pytest-html](https://pypi.org/project/pytest-html/)

This modules is a plugin for pytest that generates HTML reports for the test results. We can add this module by adding the module names & modules version into the `requirements.txt` without installing by pip manually.

## Adding to the `requirements.txt`
1. Go to plugin documentation link before
2. Obtain the name & version <br>
 <br>
3. Paste & adjust the value to the requirements
4. Select `Install requirement` from the IDE toastbar to install the modules directly
 <br>



# Generating HTML Test Report

To export thr reports for the `pytest` configuration we set before, we need to add this step:
1. Create the `reports` directory <br>

2. Navigate to the `pytest` configuration <br>

3. Modify the ***Additional Argments*** for the test configuration
    - `--html` : command to run the `pytest-html` packages
    - `reports/report.html` : define the full path for the `file directory` + `file names` that available inside cwd <br>
    
4. Run the test for the adjusted configuration <br>

5. Check the log/reports after under given path <br>

6. Open the `.html` file on the browsers by choose these options

7. Validate the results on the logs


Check the full details related the [`pytest-html`](https://pytest-html.readthedocs.io/en/latest/index.html) for the full information
